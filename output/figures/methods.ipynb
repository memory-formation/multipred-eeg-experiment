```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from matplotlib.colors import BoundaryNorm
import numpy as np

def visualize_balancing_matrix():
    # Define visual pairs with labels and weights.
    # Order: high-probability conditions first, then low-probability and neutral.
    visual_pairs = [
        ("H → CW", 3),     # predicted (high)
        ("V → CCW", 3),    # predicted (high)
        ("N → CW", 2),     # neutral
        ("N → CCW", 2),     # neutral
        ("H → CCW", 1),    # unpredicted (low)
        ("V → CW", 1)    # unpredicted (low)

    ]
    
    # Define auditory pairs with labels and weights.
    auditory_pairs = [
        ("1000Hz → 100Hz", 3),   # predicted (high)
        ("1600Hz → 160Hz", 3),   # predicted (high)
        ("N → 100Hz", 2),        # neutral
        ("N → 160Hz", 2),         # neutral
        ("1000Hz → 160Hz", 1),   # unpredicted (low)
        ("1600Hz → 100Hz", 1)   # unpredicted (low)
    ]
    
    # Create the matrix: each cell = visual weight * auditory weight.
    num_rows = len(visual_pairs)
    num_cols = len(auditory_pairs)
    matrix = np.zeros((num_rows, num_cols), dtype=int)
    
    for i, (_, v_weight) in enumerate(visual_pairs):
        for j, (_, a_weight) in enumerate(auditory_pairs):
            matrix[i, j] = v_weight * a_weight
    

    # Define the custom colormap for the visualization matrix
    colors = ['firebrick', 'grey', 'turquoise']  # Order corresponds to values 1, 3, 9
    cmap = ListedColormap(colors)

    # Define normalization to map values to colormap indices
    bounds = [0.5, 2, 4, 10]  # Bins: values close to 1, 3, 9
    norm = BoundaryNorm(bounds, cmap.N)
    # Create a heatmap of the matrix.
    fig, ax = plt.subplots(figsize=(8, 6))
    im = ax.imshow(matrix, cmap=cmap, norm=norm)
    
    # Set up ticks and labels.
    ax.set_xticks(np.arange(num_cols))
    ax.set_xticklabels([label for label, _ in auditory_pairs], rotation=45, ha="right", fontsize=10)
    ax.set_yticks(np.arange(num_rows))
    ax.set_yticklabels([label for label, _ in visual_pairs], fontsize=10)
    
    # Annotate each cell with its value.
    for i in range(num_rows):
        for j in range(num_cols):
            ax.text(j, i, matrix[i, j], ha="center", va="center", color="black", fontsize=12)
    
    ax.set_title("")
    fig.tight_layout()
    plt.savefig('methods/design_matrix.png', dpi=300)
    plt.show()
    


# save the matrix as a png
visualize_balancing_matrix()

```
